```python
thang = int(input())
nam = int(input())

if thang == 1 or thang == 3 or thang == 5 or thang == 7 or thang == 8 or thang == 10 or thang == 12:
    print(31)
elif thang == 4 or thang == 6 or thang == 9 or thang == 11:
    print(30)
elif thang == 2:
    if nam % 400 == 0 or (nam % 4 == 0 and nam % 100 != 0):
        print(29)
    else:
        print(28)
else:
    print("Tháng không hợp lệ")
```
